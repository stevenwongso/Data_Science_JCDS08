```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

### Seaborn Dataset

Testing Seaborn with their own data

seaborn dataset: https://github.com/mwaskom/seaborn-data 

seaborn documentation: https://seaborn.pydata.org/

```python
df = sb.load_dataset('flights')
# df = df[df['month'] == 'January']
# df = df[df['year'] < 1955]
df.head()
```

```python
sb.relplot(
    x='year', y='passengers', data=df, 
#     kind ='line'
    hue = 'month', size='month', 
#     style= 'year'
)
plt.yticks(rotation=90)
plt.ylabel('Penumpang')
plt.xlabel('Tahun')
plt.grid(True)
plt.title('Flights Data')
plt.show()
```

```python
bar = sb.barplot(x='year',y= 'passengers',data=df, hue='month')
# plt.legend([])  #hilangin legend
plt.legend(loc = 0)
plt.show()
```

```python
bar = sb.catplot(x='year',y= 'passengers',data=df, hue='month')
# plt.legend([])  #hilangin legend
plt.show()
```

```python
sb.heatmap(df.corr(), cmap='hot', annot=True)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sb.heatmap(df.pivot('month','year','passengers'), cmap='hot_r', annot=True,
           fmt='d' #d=int, f=float, 1f=float 1d after
          )
plt.show()
```

```python
#violin plot
sb.violinplot(x='year', y='passengers',data=df)
plt.show()
```
