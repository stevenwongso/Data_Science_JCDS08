```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,1,4,5,6,5,7,8,7]
```

```python
# plt.plot(x,y)
# plt.plot(np.array(x),np.array(y))
plt.plot(pd.Series(x),pd.Series(y))
plt.show()
```

Dapat dilakukan untuk list, numpy array, maupun pandas series

```python
df = pd.DataFrame({'X':x,'Y':y})
plt.plot(df['X'],df['Y'])
plt.title('X vs Y', fontdict={'size': 30, 'color':'red', 'family':'impact'})
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.legend(['Test 1'])
plt.grid(True)
plt.show()
```

```python
# plt.plot(df['X'],df['Y'])
# plt.plot(df['Y'],df['X'])
# plt.plot(df['X'],df['X'])
plt.plot(df['X'],df['Y'],df['Y'],df['X'],df['X'],df['X']) #one line
plt.legend(['X vs Y','Y vs X', 'X vs X'], loc=4) #bydefault location = 0 (posisi Terbaik yang tidak menghalangi)
plt.show()
```

```python
plt.plot(df['X'],df['X'], color = "red", lw=4, ls=':')
plt.plot(df['X'],df['X'] ** 2, c = 'orange', lw=3, ls='-.')
plt.plot(df['X'],df['X'] ** 3, c = '#07C092', lw=3, ls='--', marker ='$text$', ms = 40, mec='red', mfc='aquamarine')
plt.legend(['X vs X','X vs X^2', 'X vs X^3'])
plt.grid(True, color = 'aqua', linestyle='--',linewidth = 2)
plt.show()
```

```python
plt.plot(
    df['X'],df['X'], 'or-',
    df['X'],df['X']**2, 'g-.',
    df['X'],df['X']**3, 'm:'
) #one line
plt.yticks(np.arange(0,800,100),list('Aerenias'),rotation=30)
plt.xticks([0,5,10],['Andi','Budi','Caca'])
plt.savefig('mychart.png')
plt.show()
```

```python
plt.plot(df['X'], df['Y'])
plt.plot(df['X'], df['X'])
# plt.fill_between(df['X'],df['Y'],3, facecolor = 'g', alpha = 1) # angka mengindikasikan batas
plt.fill_between(df['X'],5,3, facecolor = 'pink', alpha = 0.5) # angka mengindikasikan batas
# plt.fill_between(df['X'],df['X'], facecolor = 'r', alpha = 0.7)
plt.fill_betweenx([0,8],5,3, facecolor = 'aquamarine', alpha = 0.3) # angka mengindikasikan batas
plt.show()
```

```python
plt.plot(df['X'], df['Y'], marker='o', mfc='r')
plt.fill_between(df['X'],df['Y'], facecolor = 'g', alpha = 0.3) 
plt.text(2,4, 'tes\n123', fontdict={'color':'red', 'family':'impact'})
for i in df['X'] :
    plt.text(i - 0.1 ,df['Y'].iloc[i-1] + 0.3,f"({i},{df['Y'].iloc[i-1]})" )
# plt.annotate('Highest!', xy=(8,8), xytext=(4,8), arrowprops = {'facecolor':'r','arrowstyle':'-|>'})
plt.annotate('Highest!', xy=(8,8), xytext=(4,8), arrowprops = dict(facecolor='y', shrink=.1))
plt.yticks(np.arange(0,11,2))
plt.show()
```
