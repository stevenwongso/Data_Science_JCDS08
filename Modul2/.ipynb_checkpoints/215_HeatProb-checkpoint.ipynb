## Probability

- __Sample Space__: all possible outcome
- __Outcome__ : hasil dari suatu eksperiment
- __Probability__ : chance that an event will occur (0-1)
    - __Theoretical Probability__ :
    $$P(\textrm{event}) = \frac {\textrm{jumlah outcome}}{\textrm{jumlah sample space}} = \frac{1}{2} $$
    
    - __Empirical probability__ :
        Flip coin 500x. 257 garuda. probability keluar garuda?
        
        $$ P(\textrm{garuda}) = \frac {\textrm{trial outcome garuda}}{\textrm{total trial}} = \frac {257}{500} = 0.514$$ 
        
        __Law of Large Number__ : more trials, nilai empirical probability akan mendekati theoritical probability

- Probability Formula

    - Conditional Probability $P(\textrm{A|B})$
    
        $$P(B|A) = \frac {P(A \cap B)} {P(A)} $$
    
    - AND Probability $P(\textrm{A and B})$
    
        $$P(A \cap B) =  P(A) \cdot P(B|A) $$
    
    - OR Probability $P(\textrm{A or B})$
    
        $$P(A \cup B) =  P(A) + P(B) - P(A \cap B) $$

_Example_ : Three Blue Cards  : 1 , 2 , 3. Two Yellow Cards : 1, 2. 
- $P(Blue) =\frac {3}{5}$
- $P(Yellow) = \frac {2}{5}$
- $P(Genap) = \frac {2}{5}$
- $P(Ganjil) = \frac {3}{5}$

$P(Blue|Genap) = \frac {P(Biru \cap Genap)}{P(Genap)} = \frac {\frac{1}{5}}{\frac{2}{5}}$ =0.5

$P(Biru or Genap) = P(Biru) + P(Genap) - P(Biru \cap Genap) = \frac {3}{5} + \frac {2} {5} -\frac {1}{5} $ = 0.8

Probability ambil 2x = dua-duanya = BLUE
    $$P(B1 and B2) = P(B1) \cdot P(B2) = \frac {3}{5} \cdot \frac{2}{4} = 0.3  $$ 

- __Independents Events__ : 
Syarat
    $$ P(A|B) = P(A) $$
    $$ P(B|A) = P(B) $$
    $$ P(A and B) = P(A) \cdot P(B) $$
    
    _Example_: 20 % Kidal, 5% nilai A, 1% nilai A dan Kidal
  
- P(Kidal) = 0.2
- P(Nilai A) = 0.5
- P(K and A) = 0.01

P(K and A)= P(K) * P(A) =0.01 karena memenuhi salah satu syarat dapat dikatakan bawha kidal dan nilai A independent

- __Mutually Exclusive__ : Kedua Event Tidak Bisa Bersamaan
    $$ P(A and B) = 0$$

roll a dice 1x.

A = ganjil < 4 = {1,3}

B = angka > 3 = {4,5,6}

$ P(A) = \frac {2}{6}$

$ P\textrm{(B|A)} = 0 $

$ P\textrm{(A & B)} = P(A \cdot P\textrm{(B|A)} = 0 $

A dan B tidak bisa terjadi bersamaan = Mutually exclusive

### Correlation

- __Pearson $\displaystyle r$__
- __Spearman $\displaystyle \rho$ / $r_{s}$__
- __Kendall $\displaystyle \tau $__

- Pearson Correlation Coefficient
$$ \displaystyle r = \frac {n(\sum{xy}) - (\sum{x}) (\sum{y})} {\sqrt{(n \sum{x^2}-(\sum{x})^2) (n\sum{y^2}-(\sum{y})^2)}}$$

```python
df = pd.DataFrame ({
    'mesin (x)' : [1000,2000,3000,4000,5000],
    'harga (y)' : [10,25,35,55,80]
})
df.corr(method='pearson')
```

```python
df['xy'] = df['mesin (x)'] * df ['harga (y)']
df['x2'] = df['mesin (x)'] ** 2
df['y2'] = df['harga (y)'] ** 2
n = len(df)
xy = df['xy'].sum()
x = df['mesin (x)'].sum()
y = df['harga (y)'].sum()
x2 = df['x2'].sum()
y2 = df['y2'].sum()
r = ((n * xy) - (x * y))/(np.sqrt(((n * x2) - (x**2)) * ((n * y2) - (y**2))))
r
```

- Spearman Rank $\rho$
    
    Data yang ada diranking per feature :
   - kalau ada "tied rank" : data dengan ranking yang sama
   $$\rho = \frac {\sum(x - \bar{x}) (y - \bar{y})}{\sqrt{\sum(x - \bar{x}) ^ 2 (y - \bar{y})^ 2}}$$
   
   - kalau tidak ada "tied rank"
   $$\rho = 1 - \frac {6 \sum{d}^2}{n(n^2-1)}$$
    

```python
df = pd.DataFrame({
    'Math' : [56,75,45,71,62,64,58,80,76,61],
    'Physics' : [66,70,40,60,65,56,59,77,67,63]
})
pspearman = df.corr(method='spearman')
df = df.sort_values(by='Math', ascending=False)
df['Rank Math'] = np.arange(1,11)
df = df.sort_values(by='Physics', ascending=False)
df['Rank Physics'] = np.arange (1,11)
df = df.sort_index()
#karena tidak terdapat tied rank dibutuhkan d
df['d2'] = (df['Rank Physics'] - df['Rank Math'])**2
d2 = df['d2'].sum()
n = len(df)
p = 1 - ((6 * d2) /(n*((n**2) - 1)))

display(p)
display(pspearman)
```

- Kendall $\tau$ : tiga jenis $\tau_A$, $\tau_B$, $\tau_C$, 

```python
df = df[['Math','Physics','Rank Math','Rank Physics']].copy()
df = df.sort_values(by=['Rank Math', 'Rank Physics'], ascending=True)
df['Concordant'] = np.array([9,7,7,3,1,3,2,1,1,0])
df['Disconcordant'] = np.array([0,1,0,3,4,1,1,1,0,0])

df = df.corr(method='kendall')

sumC=df['Concordant'].sum()
sumD=df['Disconcordant'].sum()
(sumC-sumD) / (sumC + sumD)


# df
```
