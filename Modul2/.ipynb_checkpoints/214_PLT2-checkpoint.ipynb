```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'Massa' : np.arange(40,100,5),
    'Tinggi' : np.arange(160,172,1)
})
df
```

```python
plt.figure('Massa vs Tinggi', figsize=(12,6))

plt.subplot(121)
plt.plot(df['Massa'], df['Tinggi'])
           
plt.subplot(122)
plt.plot(df['Tinggi'], df['Massa'])
```

```python
fig = plt.figure('Massa vs Tinggi', figsize=(12,6))

fig.add_subplot(121)
plt.plot(df['Massa'], df['Tinggi'])
           
fig.add_subplot(122)
plt.plot(df['Tinggi'], df['Massa'])
```

```python
fig,[ax1,ax2] = plt.subplots(1,2)
ax1.plot(df['Massa'], df['Tinggi'])
ax2.plot(df['Tinggi'], df['Massa'])
```

```python

ax1 = plt.subplot2grid((2,2),(1,1)) # shape(row,col), location(row,col) note untuk location index start from 0
ax1.plot(df['Massa'],df['Tinggi'])
ax2 = plt.subplot2grid((2,2),(0,0))
ax2.plot(df['Massa'],df['Tinggi'], 'r--o')
ax3 = plt.subplot2grid((2,2),(0,1))
ax3.plot(df['Massa'],df['Tinggi'], 'gD:')
ax4 = plt.subplot2grid((2,2),(1,0))
ax4.plot(df['Massa'],df['Tinggi'],'y*')
plt.show()
```

```python
plt.subplot(121)
plt.scatter(df['Massa'],df['Tinggi'], s=300, marker='*', c=list('rgbcmykrgbcm'))
plt.subplot(122)
plt.plot(df['Massa'],df['Tinggi'],'o')
```

```python
plt.subplot(121)
plt.scatter(df['Massa'],df['Tinggi'], s=300, marker='*', c=np.arange(0,1.2,0.1), cmap='PuBu', edgecolors = 'k') 
# pada cmap _r untuk reverse warna
plt.subplot(122)
plt.plot(df['Massa'],df['Tinggi'],'o')
```

```python
plt.bar(df['Massa'],df['Tinggi']) # most likely for frekuensi
plt.yticks(np.arange(150,172,10))
plt.show()
```

```python
data = [72,73,35]
notes = ['RMA','BAR', 'DRAW']
w,l,p=plt.pie(data, labels = notes, colors=['Aquamarine','purple','silver'], 
              autopct = lambda x: '{:.1f}% \n ({})'.format(x, int(round(x * sum(data)/100))), 
              shadow=True, startangle = 90, counterclock = False, explode= (0,0,0.3) ) 

#autopct =  '%1.1f%%' 1 angka di belakang koma

for i in w :
    i.set_linewidth(5)
    i.set_edgecolor('#C3FF6F')
for i in p:
    i.set_color('#3A088F')
p[1].set_color('w')
plt.show()
```

### Summarize Data Visually
- Histogram : bar , x = range value (bins), y = freq bin
- Boxplot : Q1, Q2, Q3 + outlier + min max+ whisker low outlier high outlier + data outlier

### Histogram

```python
x = [4,6,6,7,11,13,18,18,18,12,24,26,27,35,36,36,42,43,45,49]

'''
Tentukan jumlah batang yang diinginkan
lebar per bar : range/jumlah bar (45/5 = 9 ~ 10)
Tentukan bins :
        a. 3.5 - 13.5     = 6
        b. 13.5 - 23.5    = 4
        c. 23.5 - 33.5    = 3
        d. 33.5 - 43.5    = 5
        e. 43.5 - 53.5    = 2
'''
bins = [3.5,13.5,23.5,33.5,43.5,53.5]
plt.hist(x, bins, edgecolor='k', lw=5)
plt.xticks(bins)
plt.show()
```

```python
#histogram type 
# uniform : data1 , V type : data2, normal: data3 
# symmetrical:data4, skewedright:data5, skewedleft:data6

plt.figure(figsize=(12,2))
bins = [3.5,13.5,23.5,33.5,43.5,53.5]

plt.subplot(161)
data1 = [5,5,5,15,15,15,25,25,25,35,35,35,45,45,45]
plt.title('uniform')
plt.hist(data1,bins)

plt.subplot(162)
data2 = [5,5,5,5,5,15,15,25,35,35,45,45,45,45,45]
plt.title('V type')
plt.hist(data2,bins)

plt.subplot(163)
data3 = [5,5,15,15,15,15,25,25,25,25,25,25,35,35,35,35,45,45]
plt.title('normal')
plt.hist(data3,bins)

plt.subplot(164)
data4 = [5,5,5,5,15,15,25,25,25,35,35,45,45,45,45]
plt.title('symmetrical')
plt.hist(data4,bins)

plt.subplot(165)
data5 = [5,5,5,5,15,15,15,15,15,25,25,25,35,35,45]
plt.title('skewed right')
plt.hist(data5,bins)

plt.subplot(166)
data6 = [5,15,15,25,25,25,35,35,35,35,35,45,45,45,45]
plt.title('skewed left')
plt.hist(data6,bins)

plt.show()
```

```python
x = np.random.randn(10000000)
plt.hist(x)
plt.show()
```

```python
lb = (x.max() - x.min())/100
min = x.min() - (0.5*x.min())
max = x.max() + (0.5*x.min())
bins = np.arange(min,max,lb)
plt.hist(x, bins)
plt.show()
```

#### Box plot

1. Menentukan Q1, Q2, dan Q3
2. (min dan max) atau (low outlier dan high outlier berdasarkan IQR)

```python
x = [4,6,6,7,11,13,18,18,18,21,24,26,27,35,36,36,42,43,45,49,100]
df = pd.DataFrame({'X':x})
df.describe()

# Q1 = 12, Q2 = 22,5, Q3 = 36 
# min = 4 max = 49
# IQR = 24 low outlier  = -24, High Outlier = 72

# plt.boxplot(x)
plt.boxplot(x, vert=False)
plt.show()
```
