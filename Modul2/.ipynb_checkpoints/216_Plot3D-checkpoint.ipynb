```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

<hr>

### Plot 3D

```python
fig = plt.figure()
axes = plt.subplot(111, projection='3d')
x = np.arange(1,11)
y = np.arange(1,11)
z = np.arange(1,11) #2 dim
axes.plot_wireframe(x,y,np.array([z]))
axes.set_xlabel('nilai X')
axes.set_ylabel('nilai Y')
axes.set_zlabel('nilai Z')
plt.show()
```

<hr>

### Scatter 3D

```python
fig = plt.figure()
axes = plt.subplot(111, projection='3d')
x = np.arange(1,11)
y = np.arange(1,11)
z = np.arange(1,11)
axes.scatter(x,y,np.array([z]),color='r',marker='*',s=180)
plt.show()
```

### Bar 3D


axes.bar3d( titikX, titikY, titikZ  lebarX, lebarY, tinggiZ )

titik X, titik Y, titik Z sebagai alas diagram batang

titik x, titik y, dan tinggi Z merupakan value data yang ingin diplot

```python
fig = plt.figure()
axes = plt.subplot(111, projection='3d')
x = np.arange(1,11) ; y = np.arange(1,11) ; z = np.zeros(10)
lx = np.ones(10) ; ly = np.ones(10) ;tz = np.arange(1,11)
axes.bar3d(
    x,y,z,
    lx,ly,tz
)

plt.show()
```

```python
fig = plt.figure()
axes = plt.subplot(111, projection='3d')
x = np.arange(1,11) ; y = np.arange(1,11) ; z = np.arange(1,11)
lx = np.full((10,),3) ; ly = np.full((10,),0.5) ; tz = np.arange(1,11)
axes.bar3d(
    x,y,z,
    lx,ly,tz
)

plt.show()
```

```python
fig = plt.figure(figsize = (18,6))
axes = plt.subplot(121, projection='3d')
x = np.arange(1,11) ; y = np.arange(1,11) ; z = np.full((10,),8)
lx = np.full((10,),3) ; ly = np.full((10,),0.5) ; tz = np.ones(10)
axes.bar3d(
    x,y,z,
    lx,ly,tz,
    color=['r','g','b','c','m','y','k','aquamarine','silver','gold']
)
axes.set_title(label='3D Bar Plot')
axes.set_xlabel('nilai X')
axes.set_ylabel('nilai Y')
axes.set_zlabel('nilai Z')

axes = plt.subplot(122, projection='3d')
x = np.arange(1,11)
y = np.arange(1,11)
z = np.arange(1,11)
axes.scatter(x,y,np.array([z]),color='r',marker='*',s=180)

plt.show()
```
