### Pandas Series ###

```python
x = range(0,5)
xs = pd.Series(x, index=['a','b','c','d','e'], name= 'ID')
print(xs)
print(xs[2])
print(xs['b'])
print(xs.name)
```

indeks dari setiap data dapat diubah dengan menambahkan index pada perintah series
name akan menjadi nama kolom pada data frame

```python
xs = pd.Series(np.array(x))
xs
```

```python
xs = pd.Series(tuple(x))
xs
```

```python
x = {'a':0, 'b':1, "c":2, 'd':3, 'e':4, 'f':5}
xs = pd.Series(x)
xs
```

keys pada dictionary akan menjadi index dan values pada dictionary akan menjadi data

```python
xs = pd.Series('Andi')
xs
```

```python
x = [1,1,1,3,4,5]
xs = pd.Series(x, index=['a','b','c','d','e','f'], name= 'ID')

print(xs.max())
print(xs.min())
print(xs.mean())
print(xs.median())
print(xs.mode()[0])
print(xs[xs==xs.mode()[0]])
print(xs[xs==xs.mode()[0]].count)
```

Mode keluarnya dalam type series untuk mengambil valuenya dapet ditambahkan [0]

```python
print(list(xs))
```

```python
xl = list(xs)
xl = list(filter(lambda x : x == xs.mode()[0],xl))
print(xl)
print(len(xl))
```

```python
xs.tolist()
```

```python
print(xs.values.tolist())
print(xs.index.tolist())
print(xs.head(2))
print(xs.tail(2))
```

Pandas bisa langsung webscrapping dengan syarat data yang ada terdapat dalam table

```python
# install lxml html5lib
dfHtml = pd.read_html('http://digidb.io/digimon-list/')
df=dfHtml[0]
```

```python
df[df['Def'] == df['Def'].min()]
# df['HP'].max()
```

```python
df[df['Stage'] == 'Mega']['HP'].mean()
```

```python
df[df['Stage'] == 'Mega'].describe()
```

```python
# adakah digimon non mega yang hpnya berada di dalam rangenya mega
df[
    (df['Stage'] != 'Mega') 
    & 
    (df['Equip Slots'] > df[df['Stage'] == 'Mega']['Equip Slots']. min())
    & 
    (df['Equip Slots'] < df[df['Stage'] == 'Mega']['Equip Slots']. max())
]
```

```python
df[
    (df['Stage'] != 'Mega') 
    & 
    (df['Equip Slots'].between(
        df[df['Stage'] == 'Mega']['Equip Slots']. min(), 
        df[df['Stage'] == 'Mega']['Equip Slots']. max()
    ))
     ].sort_values(by='Atk')
```

```python
df[
    (df['Stage'] == 'Mega') 
    |
    (df['Stage'] == 'Baby') 
]
```

```python
df['Stage'].unique().tolist()
```

```python
df[df['Stage']=='None']
```

```python
df['Total Stat'] = df['Atk'] + df['Def'] + df['Int'] + df['Spd']
df
```

```python
df[
    (df['Stage'] != 'Mega') 
    & 
    (df['Total Stat'].between(
        df[df['Stage'] == 'Mega']['Total Stat']. min(), 
        df[df['Stage'] == 'Mega']['Total Stat']. max()
    ))
     ].sort_values(by='Total Stat')
```

```python
df['Status HP'] = df['HP'].apply(lambda z : 'High Power Hp' if z >= 700 else 'Low Power HP')
df
```

```python
print(df['Atk'].min())
print(df['Atk'].max())
print(df['Atk'].mean())

print(df['Def'].min())
print(df['Def'].max())
print(df['Def'].mean())
```

```python
# dataframes.apply(function)
df['SUPER'] = df.apply(lambda x : 'Extra Super' if x['Atk'] > df['Atk'].mean() and x['Def'] > df['Def'].mean()
                       else ('Super Attack' if (( x['Atk'] > df['Atk'].mean()) and (x['Def'] >  df['Def'].mean())) 
                             else ('Super Defense' if ( x['Atk'] < df['Atk'].mean()) and (x['Def'] >  df['Def'].mean()) 
                                   else 'Tidak Super')), axis = 1)
# df[df['SUPER'] == 'Extra Super']
df
```

```python
df['DigiStage'] = df['Digimon'] + ',' + df['Stage']
df['DigiStage'][0].split(',')
```

```python
print(df['Stage'].unique())
print(df['Type'].unique())
print(df['Attribute'].unique())
```

```python
df['Kombinasi'] = df['Stage'] + df['Type'] + df['Attribute']
print(len(df['Kombinasi'].unique()))
```
