```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_html('http://digidb.io/digimon-list/')
df = df[0]
df
```

```python
# Grouping data digimon berdasarkan kolom Type
dfg = df.groupby('Type')
```

```python
# Mengetahui ada key apa saja dalam kolom Type
dfg.groups.keys()
```

```python
# Tiap kelompok Group Type, dijadikan dataFrame sendiri-sendiri
df1 = dfg.get_group('Data')
df2 = dfg.get_group('Free')
df3 = dfg.get_group('Vaccine')
df4 = dfg.get_group('Virus')
```

```python
# Mengetahui jumlah Digimon di masing2 Type / DataFrame
print(df1['Digimon'].count())
print(df2['Digimon'].count())
print(df3['Digimon'].count())
print(df4['Digimon'].count())
```

#### CONCAT

```python
# Menggabungkan df1 dan df2
df12 = pd.concat([df1, df2])
df12

# Indexing nya masih mengikuti index di dataFrame asalnya.
```

```python
# untuk ignore index lama, dan buat index baru:
df12 = pd.concat([df1, df2], ignore_index=True)
df12
```

```python
# Untuk menampilkan nama DataFrame asal di DF baru
df12 = pd.concat([df1, df2], keys=['df1', 'df2'])
df12

# NOTE: kalau mau menampilkan dataFrame awal, tidak bisa gunakan ignore index
```

```python
# IMPORTANT!
# Kalau mau concat 2 data frame, pastikan jumlah dan jenis feature nya sama
# Kalau berbeda, dataFrame yang tidak punya feature tersebut akan muncul sebagai missing value (NaN)
```

```python
# Concat Case Example
a = pd.DataFrame([
    {'nama': 'Andi', 'usia': 12},
    {'nama': 'Budi', 'usia': 10}
])
b = pd.DataFrame([
    {'nama': 'Andi', 'massa': 30},
    {'nama': 'Budi', 'massa': 25}
]) 
```

```python
# Concat biasa
c = pd.concat([a,b], sort=False)
c

# Andi dan Budi akan muncul terpisah menjadi 4 data
```

```python
# Concat yang benar
c = pd.concat([a,b], sort=False, axis=1)
c

# sort untuk mengurutkan urutan kolom
# Data Andi dan Budi di DF1 dan DF2 akan bergabung
# CATATAN: Urutan index nama yang sama harus sama
# axis=1 artinya penggabungan DF akan row-wise (kesamping). Default concat adalah axis=0 (column-wise/kebawah)
```

```python
# Jika urutan data tidak sama:

# setting urutan indexnya
a = pd.DataFrame([
    {'nama': 'Andi', 'usia': 12},
    {'nama': 'Budi', 'usia': 10}
], index=[1,0])
b = pd.DataFrame([
    {'nama': 'Budi', 'massa': 25},
    {'nama': 'Andi', 'massa': 30}
], index=[0,1])
c = pd.concat([a,b], sort=False, axis=1)
c
```

```python
# Jika urutan data tidak sama:
# Rapikan dulu index dan urutan value nya sebelum di concat

# NOTE: Concat sama seperti stacking. Menumpuk value dari 2 sumber dataFrame.
# Dalam SQL Concat sama seperti UNION
```

#### MERGER

```python
a = pd.DataFrame([
    {'nama': 'Budi', 'usia': 12},
    {'nama': 'Caca', 'usia': 10},
    {'nama': 'Andi', 'usia': 15}
])
b = pd.DataFrame([
    {'nama': 'Caca', 'massa': 25},
    {'nama': 'Andi', 'massa': 30}
])
c = pd.merge(a,b, on='nama', how='inner')
c
```

```python
# Merger sama seperti Join dalam SQL
# on akan menjadi primary-key pencocok antara dua DF
# how akan menjadi penentu bagaimana join dilakukan. (inner, outer, right, dll)
```

#### PIVOT DATAFRAME

```python
# Pivot: Transformasi DataFrame agar tampilan lebih menarik dan mudah dipahami
a = [
    {'hari': 1, 'nama': 'Andi', 'math': 90, 'science': 80},
    {'hari': 2, 'nama': 'Andi', 'math': 65, 'science': 78},
    {'hari': 3, 'nama': 'Andi', 'math': 89, 'science': 98},
    {'hari': 1, 'nama': 'Budi', 'math': 87, 'science': 67},
    {'hari': 2, 'nama': 'Budi', 'math': 99, 'science': 90},
    {'hari': 3, 'nama': 'Budi', 'math': 78, 'science': 67},
]
a = pd.DataFrame(a)
a
```

```python
# tentukan kolom mana yang mau dijadikan index, dan kolom mana yang mau dijadikan kolom di DF baru
b = a.pivot(index='hari', columns='nama')
b

# Pivot dataframe tidak mengizinkan index yang sama. Pilih kolom yang value nya berbeda untuk dijadikan index 
```

#### PIVOT TABLE

```python
# sekilas terlihat sama
c = a.pivot_table(index='hari', columns='nama')
c
```

```python
# What exactly happen: Ada aggregate function di pivot table
# kalau indexnya sama, pivot table akan melakukan aggregasi sesuai pilihan. default: avg
a = [
    {'hari': 1, 'nama': 'Andi', 'math': 90, 'science': 80},
    {'hari': 1, 'nama': 'Andi', 'math': 65, 'science': 78},
    {'hari': 1, 'nama': 'Andi', 'math': 89, 'science': 98},
    {'hari': 1, 'nama': 'Budi', 'math': 87, 'science': 67},
    {'hari': 1, 'nama': 'Budi', 'math': 99, 'science': 90},
    {'hari': 1, 'nama': 'Budi', 'math': 78, 'science': 67},
]
a = pd.DataFrame(a)
c = a.pivot_table(index='hari', columns='nama', aggfunc='sum')
c

# aggfunc: sum, diff, mean, median, dll
```

```python
# margins=True akan memberikan tambahan kolom all yang merupakan total perkolom dan perbaris
c = a.pivot_table(index='hari', columns='nama', aggfunc='sum', margins=True)
c
```

#### MELTING

```python
a = [
    {'hari': 1, 'nama': 'Andi', 'math': 90, 'science': 80},
    {'hari': 2, 'nama': 'Andi', 'math': 65, 'science': 78},
    {'hari': 3, 'nama': 'Andi', 'math': 89, 'science': 98},
    {'hari': 1, 'nama': 'Budi', 'math': 87, 'science': 67},
    {'hari': 2, 'nama': 'Budi', 'math': 99, 'science': 90},
    {'hari': 3, 'nama': 'Budi', 'math': 78, 'science': 67},
]
a = pd.DataFrame(a)
a
```

```python
# Semua yang ada di DF berubah menjadi variabel dan value
# Cocok untuk dataframe time series
m = pd.melt(a)
m
```

```python
m = pd.melt(a, id_vars=['nama'])
m
```

```python
m = pd.melt(a, id_vars=['nama','hari'])
m
```

```python
# Penggunaan cocok untuk melting
math = [
    {'hari': 1, 'Andi': 90, 'Budi': 50},
    {'hari': 2, 'Andi': 94, 'Budi': 57},
    {'hari': 3, 'Andi': 93, 'Budi': 56}
]
math = pd.DataFrame(math)
math = pd.melt(math, id_vars=['hari'])
math

# yang di declare di id_vars akan keluar dari pasangan variable-value menjadi kolom sendiri
```

```python
# merubah nama variable-value
math = [
    {'hari': 1, 'Andi': 90, 'Budi': 50},
    {'hari': 2, 'Andi': 94, 'Budi': 57},
    {'hari': 3, 'Andi': 93, 'Budi': 56}
]
math = pd.DataFrame(math)
math = pd.melt(math, id_vars=['hari'], var_name='nama', value_name='Math')
math
```

#### CROSSTAB

```python
# Crosstab: menampilkan frequency & relation
a = [
    {'nama':'Andi', 'sex':'Male', 'kota':'Jakarta'},
    {'nama':'Budi', 'sex':'Male', 'kota':'Bandung'},
    {'nama':'Caca', 'sex':'Female', 'kota':'Jakarta'},
    {'nama':'Deni', 'sex':'Male', 'kota':'Bandung'},
    {'nama':'Euis', 'sex':'Female', 'kota':'Jakarta'},
    {'nama':'Fafa', 'sex':'Male', 'kota':'Jakarta'}
]
a = pd.DataFrame(a)
pd.crosstab(a['kota'], a['sex'])
```
