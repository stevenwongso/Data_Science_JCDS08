### 1. Toy Dataset : Boston House Prices Dataset

506 Data, 13 Features

```python
from sklearn.datasets import load_boston
```

```python
databoston = load_boston()
# databoston
```

```python
# dir(databoston) # ['DESCR', 'data', 'feature_names', 'filename', 'target']
# databoston['data'] # value dari features (X)
# databoston['data'].shape # 596 data 13 features
# databoston['data'][0] # jumlahnya ada 13 untuk merepresentasikan 13 features
databoston['feature_names']
databoston['target'][0] #mengindikasikan harga rumah dengan value data ke 0 pada 13 features
```

```python
dfBoston = pd.DataFrame(
    databoston['data'],columns = databoston['feature_names']
)
dfBoston['PRICE'] = databoston['target']
dfBoston
```

```python
# Pearson Correlation
dfBoston.corr() #untuk mendapatkan kolerasi dengan harga
# feature yang diambil : LSTAT, PTRATIO, TAX, RM, INDUS (berdasarkan kolerasi)
```

```python
# dapat juga dilakukan visualisasi data terlebih dahulu

import seaborn as sb
sb.pairplot(dfBoston)
```

```python
from sklearn.linear_model import LinearRegression
modelboston = LinearRegression()
```

```python
# training misalkan untuk all data dengan feature yang diambil : LSTAT, PTRATIO, TAX, RM, INDUS
modelboston.fit(dfBoston[['LSTAT', 'PTRATIO', 'TAX', 'RM', 'INDUS']], dfBoston['PRICE'])
```

```python
display(modelboston.coef_)
display(modelboston.intercept_)
dfBoston['PRICEpred'] = modelboston.predict(dfBoston[['LSTAT', 'PTRATIO', 'TAX', 'RM', 'INDUS']])
dfBoston
```

```python
dfBoston['PRICEpred'][dfBoston['PRICEpred'] <0] = 0
```

```python
from sklearn.metrics import(
    max_error,
    mean_absolute_error,
    mean_squared_error,
    mean_squared_log_error,
    median_absolute_error,
    r2_score
)
```

```python
print(f"Max Error = {max_error(dfBoston['PRICE'],dfBoston['PRICEpred'])}")
print(f"MAE = {mean_absolute_error(dfBoston['PRICE'],dfBoston['PRICEpred'])}")
print(f"MSE = {mean_squared_error(dfBoston['PRICE'],dfBoston['PRICEpred'])}")
print(f"RMSE = {np.sqrt(mean_squared_error(dfBoston['PRICE'],dfBoston['PRICEpred']))}")
print(f"MSLE = {mean_squared_log_error(dfBoston['PRICE'],dfBoston['PRICEpred'])}")
print(f"RMSLE = {np.sqrt(mean_squared_log_error(dfBoston['PRICE'],dfBoston['PRICEpred']))}")
print(f"Median Absolute Error = {median_absolute_error(dfBoston['PRICE'],dfBoston['PRICEpred'])}")
print(f"R2 = {r2_score(dfBoston['PRICE'],dfBoston['PRICEpred'])}")
```

```python
dfBoston.describe()
```

### 3. Real World Datasets : California House Price Datasets

```python
from sklearn.datasets import fetch_california_housing
```

```python
datacali= fetch_california_housing()
```

```python
# dir(datacali) # ['DESCR', 'data', 'feature_names', 'target']
# datacali['data'].shape # 20640 baris data dengan 8 feature X / colom
# datacali['feature_names']
# datacali['target'].shape 
```

```python
dfcali= pd.DataFrame(datacali['data'], columns = datacali['feature_names'])
dfcali['Price'] = datacali['target']
dfcali.head()
```

```python
display(dfcali.describe())
display(dfcali.isnull().sum())
```

```python
dfcali.corr()
# walaupun yang mempengaruhi kuat hanya MedInc namun digunakan  4 feature 'MedInc','HouseAge','AveRooms','AveBedrms ' dengan alasan logika
```

```python
modelCali = LinearRegression()
```

```python
modelCali.fit(dfcali[['MedInc','HouseAge','AveRooms','AveBedrms']],dfcali['Price'])
```

```python
dfcali['PricePred'] = modelCali.predict(dfcali[['MedInc','HouseAge','AveRooms','AveBedrms']])
dfcali
```

```python
dfcali['PricePred'][dfcali['PricePred'] <0] = 0
```

```python
print(f"Max Error = {max_error(dfcali['Price'],dfcali['PricePred'])}")
print(f"MAE = {mean_absolute_error(dfcali['Price'],dfcali['PricePred'])}")
print(f"MSE = {mean_squared_error(dfcali['Price'],dfcali['PricePred'])}")
print(f"RMSE = {np.sqrt(mean_squared_error(dfcali['Price'],dfcali['PricePred']))}")
print(f"MSLE = {mean_squared_log_error(dfcali['Price'],dfcali['PricePred'])}")
print(f"RMSLE = {np.sqrt(mean_squared_log_error(dfcali['Price'],dfcali['PricePred']))}")
print(f"Median Absolute Error = {median_absolute_error(dfcali['Price'],dfcali['PricePred'])}")
print(f"R2 = {r2_score(dfcali['Price'],dfcali['PricePred'])}")
```
