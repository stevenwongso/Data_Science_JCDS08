# Regularization

Ridge, LASSO & Elastic-Net Regression

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
df = pd.DataFrame({
    'x': np.arange(10),
    'y': np.arange(10) + np.random.randn(10)
})
# df
````

````python
plt.plot(df['x'], df['y'], 'ro')
plt.show()
````

````python
from sklearn.linear_model import LinearRegression
````

````python
model = LinearRegression()
````

````python
# training
model.fit(df[['x']], df['y'])
````

````python
model.coef_[0], model.intercept_
````

````python
# akurasi prediksi thd x & y yg dijadikan sbg data training
# akurasi : seberapa sering tebakan model benar?
model.score(df[['x']], df['y'])
````

````python
plt.plot(df['x'], df['y'], 'ro')
plt.plot(df['x'], model.coef_[0] * df['x'] + model.intercept_, 'g-')
plt.show()
````

````python
model2 = LinearRegression()
````

````python
# model2 ditrain dg hny 2 data pertama
model2.fit(df.iloc[:2][['x']], df.iloc[:2]['y'])
````

````python
plt.plot(df['x'], df['y'], 'ro')
plt.plot(df['x'], model.coef_[0] * df['x'] + model.intercept_, 'g-')
plt.plot(df['x'], model2.coef_[0] * df['x'] + model2.intercept_, 'y-')
plt.show()
````

# Ridge Regression

- L2 regularization SSE = SSE $\displaystyle + \lambda \cdot m^2$
- Di Sklearn besarnya penalty bisa diatur melalui property ```alpha```
- Jika ```alpha = 0``` maka hasilnya = simple linear regression

````python
from sklearn.linear_model import Ridge
````

````python
# train dg 2 dataset pertama
model3 = Ridge(alpha=0.75)
model3.fit(df.iloc[:2][['x']], df.iloc[:2]['y'])
````

````python
plt.plot(df['x'], df['y'], 'ro')
plt.plot(df['x'], model.coef_[0] * df['x'] + model.intercept_, 'g-')
plt.plot(df['x'], model2.coef_[0] * df['x'] + model2.intercept_, 'y-')
plt.plot(df['x'], model3.predict(df[['x']]), 'c-')
plt.show()
````

# LASSO Regression

- LASSO (Least absolute shrinkage & selection operator)
- L1 norm regularization SSE = SSE $\displaystyle + \lambda \cdot |m|$
- Di sklearn nilai ralat/penalty diatur dengan property ```alpha```
- Jika ```alpha = 0``` maka hasilnya sama seperti linear regression biasa

````python
from sklearn.linear_model import Lasso
````

````python
model4 = Lasso(alpha=0.3)
````

````python
# train hny dg 2 data pertama
model4.fit(df.iloc[:2][['x']], df.iloc[:2]['y'])
````

````python
plt.plot(df['x'], df['y'], 'ro')
plt.plot(df['x'], model.coef_[0] * df['x'] + model.intercept_, 'g-')
plt.plot(df['x'], model2.coef_[0] * df['x'] + model2.intercept_, 'y-')
plt.plot(df['x'], model3.predict(df[['x']]), 'c-')
plt.plot(df['x'], model4.predict(df[['x']]), 'm-')
plt.show()
````

````python
print(model.coef_)
print(model2.coef_)
print(model3.coef_)
print(model4.coef_)
````

# Elastic-Net

- L1 + L2 regularization
- SSE = SSE $\displaystyle + \lambda_{\textrm{ridge}} \cdot m^2 + \lambda_{\textrm{lasso}} \cdot |m|$
- Di sklearn, nilai penalty bisa diatur via prop ```alpha```

````python
from sklearn.linear_model import ElasticNet
model5 = ElasticNet(alpha=0.35)
model5.fit(df.iloc[:2][['x']], df.iloc[:2]['y'])
````

````python
plt.plot(df['x'], df['y'], 'ro')
plt.plot(df['x'], model.coef_[0] * df['x'] + model.intercept_, 'g-')
plt.plot(df['x'], model2.coef_[0] * df['x'] + model2.intercept_, 'y-')
plt.plot(df['x'], model3.predict(df[['x']]), 'c-')
plt.plot(df['x'], model4.predict(df[['x']]), 'm-')
plt.plot(df['x'], model5.predict(df[['x']]), 'r-')
plt.show()
````
