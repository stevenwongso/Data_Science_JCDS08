# Feature Scalling

### 1. Maximum Absolute Scaler


- Maximum Absolute Scaler formula $\displaystyle x' = \frac {x} {|x|_\textrm{max}}$

- Misal $x$ = [1, 2, -4] maka $|x|_\textrm{max} = 4$ sehingga $x'$ = [1/4, 2/4, -4/4]

- Data hasil MaxAbsScaler memiliki range __-1__ sampai __1__.

````python
# maximum absolute
x = [1,2,-4]
np.max(np.abs(x))
````

````python
df = pd.DataFrame({
    'TB': [1.72, 1.81, 1.93, 1.67, 1.85, 1.66, 1.59, 1.76, 1.88, 1.78],
    'BB': [76, 65, 87, 55, 60, 78, 98, 77, 74, 64]
})
df.head(2)
````


### A. Manual

````python
# 1. manual calculation
df['masTB'] = df['TB'] / np.max(np.abs(df['TB']))
df['masBB'] = df['BB'] / np.max(np.abs(df['BB']))
df
````


### B. Sklearn ```maxabs_scale()```

````python
from sklearn.preprocessing import maxabs_scale
````

````python
df['masTB2'] = maxabs_scale(df['TB'])
df['masBB2'] = maxabs_scale(df['BB'])
df.head()
````


### C. Sklearn ```MaxAbsScaler()```

````python
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
````

````python
scaler.fit(df[['TB', 'BB']])
mas = scaler.transform(df[['TB','BB']])
df['masTB3'] = mas[:,0]
df['masBB3'] = mas[:,1]
df.head()
````

````python
# inverse transform
inmas = scaler.inverse_transform(df[['masTB3','masBB3']])
df['inmasTB3'] = inmas[:,0]
df['inmasBB3'] = inmas[:,1]
df.head()
````

<hr>

#### Apakah MaxAbsScaler merubah sebaran data?

Tidak

````python
plt.subplot(121)
plt.plot(df['TB'], df['BB'], 'r.')
plt.subplot(122)
plt.plot(df['masTB'], df['masBB'], 'g.')
plt.show()
````

# Feature Scalling

### 2. Robust Scaller


- Robust scaler formula $\displaystyle x' = \frac {x - \textrm{median}(x)} {\textrm{IQR}} = \frac {x - Q_2} {Q_3 - Q_1}$

- Bagus digunakan pada data dengan banyak outliers.

````python
df = pd.DataFrame({
    'TB': [1.72, 1.81, 1.93, 1.67, 1.85, 1.66, 1.59, 1.76, 1.88, 1.78],
    'BB': [76, 65, 87, 55, 60, 78, 98, 77, 74, 64]
})
````


### A. Manual Calculation

````python
q1TB = np.quantile(df['TB'], .25)
q2TB = np.quantile(df['TB'], .5)
q3TB = np.quantile(df['TB'], .75)

q1BB = np.quantile(df['BB'], .25)
q2BB = np.quantile(df['BB'], .5)
q3BB = np.quantile(df['BB'], .75)

display( q1TB,q2TB,q3TB)
display( q1BB,q2BB,q3BB)
````

````python
df['rsTB'] = (df['TB'] - q2TB) / (q3TB - q1TB)
df['rsBB'] = (df['BB'] - q2BB) / (q3BB - q1BB)
df
````

### B. Sklearn ```RobustScaler()```

````python
from sklearn.preprocessing import RobustScaler
````

````python
scaler = RobustScaler()
scaler.fit(df[['TB', 'BB']])
rs = scaler.transform(df[['TB', 'BB']])

df['rsTB2'] = rs[:, 0]
df['rsBB2'] = rs[:, 1]
df
````

````python
# mengembalikan ke value semula / inverse transform
inrs = scaler.inverse_transform(df[['rsTB2', 'rsBB2']])
df['inrsTB'] = inrs[:, 0]
df['inrsBB'] = inrs[:, 1]
df
````

<hr>

#### Apakah Robust Scaler merubah sebaran data?

Tidak

````python
plt.subplot(121)
plt.plot(df['TB'], df['BB'], 'r.')
plt.subplot(122)
plt.plot(df['rsTB'], df['rsBB'], 'g.')
plt.show()
````

# Transformer

### 3. Binarizer

- Merubah data menjadi data biner (value 0 atau 1) aka boolean dengan menggunakan threshold tertentu. Misal diambil threshold = 20, maka data dengan value <= 20 akan berubah jadi 0 & data value > 20 akan menjadi 1.

````python
df = pd.DataFrame({
    'TB': [1.72, 1.81, 1.93, 1.67, 1.85, 1.66, 1.59, 1.76, 1.88, 1.78],
    'BB': [76, 65, 87, 55, 60, 78, 98, 77, 74, 64]
})
````

### A. Manual Calculation

Set threshold TB = 1.80 & threshold BB = 80

````python
df['bTB'] = df['TB'].apply(lambda x: 0 if x <= 1.8 else 1)
df['bBB'] = df['BB'].apply(lambda x: 0 if x <= 80 else 1)
df
````

### B. Sklearn ```Binarizer```

Set threshold TB = 1.80 & threshold BB = 80

````python
from sklearn.preprocessing import Binarizer
````

````python
transformer = Binarizer(threshold=1.8)
transformer.fit(df[['TB']])
df['bTB2'] = transformer.transform(df[['TB']]).astype('int32')
````

````python
transformer = Binarizer(threshold=80)
transformer.fit(df[['BB']])
df['bBB2'] = transformer.transform(df[['BB']])
df
````

# Transformer

### 5. Normalize

- Normaliation : Scale input vectors individually to unit form
- Scalling inputs to unit norms is a common operation for text classification or clustering
- Defaultnya perlakuannya per baris

````python
df = pd.DataFrame({
    'TB': [1.72, 1.81, 1.93, 1.67, 1.85, 1.66, 1.59, 1.76, 1.88, 1.78],
    'BB': [76, 65, 87, 55, 60, 78, 98, 77, 74, 64]
})
````

````python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
ss = scaler.fit_transform(df[['TB', 'BB']])
df['ss TB'] = ss[:, 0]
df['ss BB'] = ss[:, 1]
df
````

### A. Manual Calculation

- ada 3 teknik normalize :
    - Normalize Max $ \displaystyle x' = \frac {x}{x_\textrm{max}}$
    - Normalize L1 $ \displaystyle x' = \frac {x}{\sum x}$
    - Normalize L2 $ \displaystyle x' = \frac {x}{\sqrt{\sum x^2}}$

````python
# 1. Max Normalize

x = []
for i in range(len(df)):
    data = df.iloc[i][['ss TB', 'ss BB']] / df.iloc[i][['ss TB', 'ss BB']].max()
    x.append(data.tolist())
x = np.array(x)

df['ssTB nmax'] = x[:, 0]
df['ssBB nmax'] = x[:, 1]
df
````

````python
# 2. L1 Normalize

x = []
for i in range(len(df)):
    data = df.iloc[i][['ss TB', 'ss BB']] / df.iloc[i][['ss TB', 'ss BB']].sum()
    x.append(data.tolist())
x = np.array(x)

df['ssTB nL1'] = x[:, 0]
df['ssBB nL1'] = x[:, 1]
df
````

````python
# 3. L2 Normalize

x = []
for i in range(len(df)):
    data = df.iloc[i][['ss TB', 'ss BB']] / np.sqrt(np.sum(df.iloc[i][['ss TB', 'ss BB']] ** 2))
    x.append(data.tolist())
x = np.array(x)

df['ssTB nL2'] = x[:, 0]
df['ssBB nL2'] = x[:, 1]
df
````

### B. Sklearn ```normalize()```

````python
df = pd.DataFrame({
    'TB': [1.72, 1.81, 1.93, 1.67, 1.85, 1.66, 1.59, 1.76, 1.88, 1.78],
    'BB': [76, 65, 87, 55, 60, 78, 98, 77, 74, 64]
})

scaler = StandardScaler()
ss = scaler.fit_transform(df[['TB', 'BB']])
df['ss TB'] = ss[:, 0]
df['ss BB'] = ss[:, 1]

df
````

````python
from sklearn.preprocessing import normalize
````

````python
nmax = normalize(df[['ss TB', 'ss BB']], norm='max')
nL1 = normalize(df[['ss TB', 'ss BB']], norm='l1')
nL2 = normalize(df[['ss TB', 'ss BB']], norm='l2')

df['ssTB nmax'] = nmax[:, 0]
df['ssBB nmax'] = nmax[:, 1]
df['ssTB nL1'] = nL1[:, 0]
df['ssBB nL1'] = nL1[:, 1]
df['ssTB nL2'] = nL2[:, 0]
df['ssBB nL2'] = nL2[:, 1]

df
````

### C. Sklearn ```Normalizer()```

````python
df = pd.DataFrame({
    'TB': [1.72, 1.81, 1.93, 1.67, 1.85, 1.66, 1.59, 1.76, 1.88, 1.78],
    'BB': [76, 65, 87, 55, 60, 78, 98, 77, 74, 64]
})

scaler = StandardScaler()
ss = scaler.fit_transform(df[['TB', 'BB']])
df['ss TB'] = ss[:, 0]
df['ss BB'] = ss[:, 1]

df
````
