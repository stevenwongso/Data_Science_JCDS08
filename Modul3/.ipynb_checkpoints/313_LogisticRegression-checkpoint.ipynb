````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

# Logistic Regression

- Machine Learning for Classification Cases especially for binary classification
- Formula $\displaystyle \hat{y} = \frac {1}{1 + e^{-(mx+c)}}$ dengan e bilangan Euler : 2.718

````python
# data hasil promo kartu kredit terhadap pelanggan usia tertentu
df = pd.DataFrame({
    'usia': range (25,55),
})
df['beli'] = df['usia'].apply(lambda x: 1 if x >= 40 else 0)
df.head()
````

````python
plt.plot(df['usia'],df['beli'],'ro')
plt.show()
````

````python
# muncul logistic regression akibat binary, jadi terdapat dua nilai 1, 0
# menggunakan threshold (default 0.5)
# hasil dari logistic merupakan nilai probability
# probability < Threshold maka predict = 0
# probability > threshold maka predict = 1


# Bandingkan Linear Regression dan Logistic Regression
from sklearn.linear_model import LogisticRegression,LinearRegression
modelA = LinearRegression()
modelB = LogisticRegression()
````

````python
modelA.fit(df[['usia']],df['beli'])
modelB.fit(df[['usia']],df['beli'])
````

````python
# Gradient/Slope/M & Intercept.c Model LinReg
display('ModelA : Linear Regression')
display(modelA.coef_[0],modelA.intercept_)

# Gradient/Slope/M & Intercept.c Model LogReg
display('ModelB : Logistic Regression')
display(modelB.coef_[0,0],modelB.intercept_[0])
````

````python
# best fit line model A

df['predA'] = modelA.predict(df[['usia']])
df.head()

# bisa dialter juga misal dengan mendifinisikan threshold yang ada : contoh kalau < 0.5 tidak beli (0) dan > 0.5 beli (1)
````

````python
# best fit line model B

df['predB'] = modelB.predict(df[['usia']])
df['probabilityB'] = 1/(1 + np.exp(-1 *((modelB.coef_[0,0] * df['usia'] )+ modelB.intercept_[0])))
df
````

````python
# visualisasi best fit line model A dan B

plt.plot(df['usia'],df['beli'],'r.')
plt.plot(df['usia'],df['predA'],'g-')
plt.plot(df['usia'],df['probabilityB'],'m-')
plt.xlabel('Usia') ; plt.ylabel('y Probability')
plt.show()
````

````python
# Probability
modelB.predict_proba(df[['usia']])

# interpretasi : contoh baris pertama
# [9.99999970e-01, 3.00531898e-08] : 0.99 tidak beli (0) & 3. 10^-8 beli (1) 

df['probaBtidakbeli'] = modelB.predict_proba(df[['usia']])[:,0]
df['probaBbeli'] = modelB.predict_proba(df[['usia']])[:,1]
df
````

````python
modelB.score(df[['usia']],df['beli'])
````

````python
# mirip dengan score hanya ini lebih seperti membandingkan antara prediksi dengan data target asli
from sklearn.metrics import accuracy_score
````

````python
print(accuracy_score(df['beli'], modelB.predict(df[['usia']])))
````

### Contoh kasus :  ``` load_iris ```
Iris dataset Classification

````python
from sklearn.datasets import load_iris
dataIris = load_iris()
````

````python
dfIris = pd.DataFrame(
    dataIris['data'],
    columns = ['SL','SW','PL','PW'] # dataIris['feature_names']
)
dfIris['target'] = dataIris['target']
dfIris['spesies'] = dfIris['target'].apply( lambda x : dataIris['target_names'][x])
dfIris.head()
````

````python
# Train Test Split  training 80$ testing 20%
from sklearn.model_selection import train_test_split

xtr, xts, ytr, yts = train_test_split(dfIris[['SL','SW','PL','PW']],dfIris['spesies'], test_size = 0.2)
````

````python
xtr
````

````python
modelLR= LogisticRegression(multi_class='auto', solver='liblinear') 
# kalau mau logistic Regression untuk multi class lebih baik tambah param multi class auto
# solver lebih ke teknik penyelesaiin default lblgs : namun dalam kasus ini iterasi kurang dan gradien belum optimal , baca dokumentasi untuk setiap solver
````

````python
modelLR.fit(xtr,ytr)
````

````python
dfIris['prediksi'] = modelLR.predict(dfIris[['SL','SW','PL','PW']])
dfIris
# ketika multi class : setosa & bukan sentosa, versicolor & bukan versicolor, virginica & bukan virginica
# lalu dibandingkan nilainya
````

````python
# Prediksi Meleset
dfIris[dfIris['spesies']!=dfIris['prediksi']]
````

````python
print(modelLR.score(xtr,ytr))
print(modelLR.score(xts,yts))
````

````python
# cek probability tiap data
proba = modelLR.predict_proba(dfIris[['SL','SW','PL','PW']])
dfIris['probaSetosa'] = proba[:,0]
dfIris['probaVersicolor'] = proba[:,1]
dfIris['probaVirginica'] = proba[:,2]
dfIris

# dapat disimpulkan nilai probab terbesar yang menjadi hasil prediksi
````
