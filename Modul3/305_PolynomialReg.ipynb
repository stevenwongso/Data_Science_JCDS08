# Polynomial Regression

- Polynomial regression $\displaystyle y = m_1 x + m_2 x^2 + m_3 x^3 + ... + m_n x^n + c$
- Polynomial regression sebenarnya hanyalah multiple linear regression dengan feature x beserta nilai polinomialnya hingga n tertentu: ${x, x^2, x^3, ... x^n}$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,721,10) * np.pi/180
y = np.sin(x)
df = pd.DataFrame({
    'x': x, 'y': y
})
```

```python
plt.plot(df['x'], df['y'], 'r.')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
```

```python
modelA = LinearRegression()
modelB = Ridge(alpha=1)
modelC = Lasso(alpha=1)
modelD = ElasticNet(alpha=1)
```

```python
modelA.fit(df[['x']], df['y'])
modelB.fit(df[['x']], df['y'])
modelC.fit(df[['x']], df['y'])
modelD.fit(df[['x']], df['y'])
```

```python
plt.plot(df['x'], df['y'], 'r.')
plt.plot(df['x'], modelA.predict(df[['x']]), 'g-')
plt.plot(df['x'], modelB.predict(df[['x']]), 'b-')
plt.plot(df['x'], modelC.predict(df[['x']]), 'y-')
plt.plot(df['x'], modelD.predict(df[['x']]), 'c-')
plt.show()
```

```python
# Polynomial regression
for i in range(2, 11):
    df[f'x{i}'] = df['x'] ** i
```

### Polynomial Feature with Sklearn

```python
from sklearn.preprocessing import PolynomialFeatures
x = np.arange(1,4)
x
```

```python
# poly = transformer: mengubah bentuk data

poly = PolynomialFeatures(3, include_bias=False) 
# polinomial sampai pangkat 3
# include_bias=False ~ pangkat mulai dari pangkat 1!

poly.fit(x.reshape(-1,1)) # create transformer
# kenapa reshape? krn .fit butuh array 2D
poly.transform(x.reshape(-1,1)) # mengubah data dg transformer
```

```python
poly.fit_transform(x.reshape(-1, 1))
```
